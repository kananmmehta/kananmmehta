```python
import os
from stable_baselines3 import A2C
import gymnasium as gym
from stable_baselines3.common.env_util import make_atari_env
from stable_baselines3.common.vec_env import VecFrameStack, VecVideoRecorder # Re-import VecFrameStack
from IPython.display import HTML
from base64 import b64encode

# Re-create the environment for the model with consistent frame stacking
env_test = make_atari_env(environment_name, n_envs=1, seed=0)
env_test = VecFrameStack(env_test, n_stack=4) # Apply VecFrameStack here

# Load the saved model
a2c_path = os.path.join('Training','Saved Models','A2C_Breakout_Model')
model = A2C.load(a2c_path, env=env_test)

# Create a video recording environment with consistent frame stacking
env_record = make_atari_env(environment_name, n_envs=1, seed=0)
env_record = VecFrameStack(env_record, n_stack=4) # Apply VecFrameStack here

video_folder = 'videos'
video_length = 5000  # Number of timesteps to record

env_video = VecVideoRecorder(env_record, video_folder,
                               record_video_trigger=lambda x: x == 0,
                               video_length=video_length,
                               name_prefix=f"a2c_breakout")

# Reset the environment and run the agent
obs = env_video.reset()
for i in range(video_length):
    action, _states = model.predict(obs, deterministic=True)
    obs, rewards, dones, info = env_video.step(action)
    if dones.any():
        break

env_video.close()

# Function to display video
def show_videos(video_path='', prefix=''):
    html = []
    # Ensure the video folder exists before trying to list its contents
    os.makedirs(video_path, exist_ok=True)
    for mp4 in os.listdir(video_path):
        if os.path.isfile(os.path.join(video_path, mp4)) and mp4.startswith(prefix):
            html.append(f"<video width='400' controls><source src='{os.path.join(video_path, mp4)}' type='video/mp4'></video>")
    return HTML("<br>".join(html))


# Display the recorded video
print(f"Displaying video from: {video_folder}")
# This line is important to ensure the directory exists for os.listdir
os.makedirs(video_folder, exist_ok=True)
show_videos(video_folder, prefix='a2c_breakout')
```
